```python
from basketball_reference_web_scraper import client
from basketball_reference_web_scraper.data import OutputType
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#gets player stat totals per season
def get_player_season_stats(year):
    client.players_season_totals(
        season_end_year= year, 
        output_type=OutputType.CSV, 
        output_file_path=f"./total_stats/{year-1}_{year}_player_season_totals.csv"
)
```

```python
#gets player advanced stat totals per season
def get_player_advanced_season_stats(year):
    client.players_advanced_season_totals(
        season_end_year=year,
        output_type=OutputType.CSV,
        output_file_path=f"./advanced_season_stat_total/{year-1}_{year}_advanced_player_season_totals.csv"
)
```

```python
#gets player per game stats per season
def get_per_game_stat(year):
    req = requests.get(f'https://www.basketball-reference.com/leagues/NBA_{year}_per_game.html')
    soup = BeautifulSoup(req.content, 'html.parser')
    per_game_stat = []
    #finding each player and their specific stat
    for stat in soup.find('tbody').find_all('tr'):
        per_game_stat_dict = {}
        #find player name
        player = stat.find('td', {'data-stat':'player'})
        #need to add if statement so it only recognizes player names that are filled out
        if player:
            player_found = player.find('a')
        per_game_stat_dict['player'] = player_found.getText()
    
        #finds position
        position = stat.find('td', {'data-stat':'pos'})
        if position:
            per_game_stat_dict[f'position_{year-1}_{year}'] = position.getText()
        
        #find age    
        age = stat.find('td', {'data-stat':'age'})
        if age:
            per_game_stat_dict[f'age_{year-1}_{year}'] = age.getText()
     
        #find team id
        team_id = stat.find('td', {'data-stat':'team_id'})
        if team_id:
            per_game_stat_dict[f'team_id_{year-1}_{year}'] = team_id.getText()
    
        #find games played
        games = stat.find('td', {'data-stat':'g'})
        if games:
            per_game_stat_dict[f'games_played_{year-1}_{year}'] = games.getText()
    
        #find games started
        games_start = stat.find('td', {'data-stat':'gs'})
        if games_start:
            per_game_stat_dict[f'games_started_{year-1}_{year}'] = games_start.getText()
        
        #find mpg
        mpg = stat.find('td', {'data-stat':'mp_per_g'})
        if mpg:
            per_game_stat_dict[f'minutes_per_game_{year-1}_{year}'] = mpg.getText()
        
        #find fgm per game
        fgm = stat.find('td', {'data-stat':'fg_per_g'})
        if fgm:
            per_game_stat_dict[f'fgm_per_game_{year-1}_{year}'] = fgm.getText()
        
        #find fga per game
        fga = stat.find('td', {'data-stat':'fga_per_g'})
        if fga:
            per_game_stat_dict[f'fga_per_game_{year-1}_{year}'] = fga.getText()
    
        #find fgm per game
        fg_pct = stat.find('td', {'data-stat':'fg_pct'})
        if fg_pct:
            per_game_stat_dict[f'fg_pct_per_game_{year-1}_{year}'] = fg_pct.getText()
        
        #find 3pt fgm per game
        fg3m = stat.find('td', {'data-stat':'fg3_per_g'})
        if fg3m:
            per_game_stat_dict[f'3pt_fgm_per_game_{year-1}_{year}'] = fg3m.getText()
        
         #find 3pt fga per game
        fga3 = stat.find('td', {'data-stat':'fg3a_per_g'})
        if fga3:
            per_game_stat_dict[f'3pt_fga_per_game_{year-1}_{year}'] = fga3.getText()
        
        #find 3pt pct per game
        fg3_pct = stat.find('td', {'data-stat':'fg3_pct'})
        if fg3_pct:
            per_game_stat_dict[f'3pt_fg_pct_per_game_{year-1}_{year}'] = fg3_pct.getText()
        
        #find 2pt fgm per game
        fg2m = stat.find('td', {'data-stat':'fg2_per_g'})
        if fg2m:
            per_game_stat_dict[f'2pt_fgm_per_game_{year-1}_{year}'] = fg2m.getText()
    
        #find 2pt fga per game
        fg2a = stat.find('td', {'data-stat':'fg2a_per_g'})
        if fg2a:
            per_game_stat_dict[f'2pt_fga_per_game_{year-1}_{year}'] = fg2a.getText()
        
        #find 2pt fg pct per game
        fg2_pct = stat.find('td', {'data-stat':'fg2_pct'})
        if fg2_pct:
            per_game_stat_dict[f'2pt_fg_pct_per_game_{year-1}_{year}'] = fg2_pct.getText()
        
        #find ef fg pct per game
        ef_fg_pct = stat.find('td', {'data-stat':'efg_pct'})
        if ef_fg_pct:
            per_game_stat_dict[f'effective_fg_pct_per_game_{year-1}_{year}'] = ef_fg_pct.getText()
        
        #find ftm per game
        ftm = stat.find('td', {'data-stat':'ft_per_g'})
        if ftm:
            per_game_stat_dict[f'ftm_per_game_{year-1}_{year}'] = ftm.getText()
        
        #find fta per game
        fta = stat.find('td', {'data-stat':'fta_per_g'})
        if fta:
            per_game_stat_dict[f'fta_per_game_{year-1}_{year}'] = fta.getText()
        
        #find ft pct per game
        ft_pct = stat.find('td', {'data-stat':'ft_pct'})
        if ft_pct:
            per_game_stat_dict[f'ft_pct_per_game_{year-1}_{year}'] = ft_pct.getText()
        
        #find oreb per game
        oreb = stat.find('td', {'data-stat':'orb_per_g'})
        if oreb:
            per_game_stat_dict[f'oreb_per_game_{year-1}_{year}'] = oreb.getText()
        
        #find dreb per game
        dreb = stat.find('td', {'data-stat':'drb_per_g'})
        if dreb:
            per_game_stat_dict[f'dreb_per_game_{year-1}_{year}'] = dreb.getText()
        
        #find reb per game
        reb = stat.find('td', {'data-stat':'trb_per_g'})
        if reb:
            per_game_stat_dict[f'reb_per_game_{year-1}_{year}'] = reb.getText()
            
        #find ast per game
        ast = stat.find('td', {'data-stat':'ast_per_g'})
        if ast:
            per_game_stat_dict[f'ast_per_game_{year-1}_{year}'] = ast.getText()
        
        #find stl per game
        stl = stat.find('td', {'data-stat':'stl_per_g'})
        if stl:
            per_game_stat_dict[f'stl_per_game_{year-1}_{year}'] = stl.getText()

        #find blk per game
        blk = stat.find('td', {'data-stat':'blk_per_g'})
        if blk:
            per_game_stat_dict[f'blk_per_game_{year-1}_{year}'] = blk.getText()
    
        #find tov per game
        tov = stat.find('td', {'data-stat':'tov_per_g'})
        if tov:
            per_game_stat_dict[f'tov_per_game_{year-1}_{year}'] = tov.getText()
    
        #find pf per game
        pf = stat.find('td', {'data-stat':'pf_per_g'})
        if pf:
            per_game_stat_dict[f'pf_per_game_{year-1}_{year}'] = pf.getText()
        
        #find ppg
        ppg = stat.find('td', {'data-stat':'pts_per_g'})
        if ppg:
            per_game_stat_dict[f'ppg_{year-1}_{year}'] = ppg.getText()
        
    #for cat in stat.find('td', {'data-stat': cat}):
     #   if cat:
     #       per_game_stat_dict[cat] = cat.getText()
    
        per_game_stat.append(per_game_stat_dict)
    df = pd.DataFrame(per_game_stat)    
    return df.to_csv(f'./per_game_stat_csv/per_game_stat_{year-1}_{year}.csv')
```
